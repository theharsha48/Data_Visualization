 3. Create a dataset representing categories and their corresponding values. Compare different categories 
based on numerical values. 

```python
import pandas as pd

# Create a dataset representing categories and their corresponding values
data = {
    'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [10, 20, 30, 15, 25, 35, 12, 22, 32]
}
df = pd.DataFrame(data)

# Group the data by category and calculate the mean value for each category
category_means = df.groupby('Category')['Value'].mean()

# Compare different categories based on numerical values
print("Mean values by category:")
print(category_means)

# Visualize the comparison using a bar chart
plt.bar(category_means.index, category_means.values)
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.title('Comparison of Categories by Mean Value')
plt.show()
```

 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical 
variable across different categories.

```python
import seaborn as sns

# Generate a dataset with categories and numerical values
np.random.seed(0)
data = {
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value': np.random.normal(0, 1, 100)
}
df = pd.DataFrame(data)

# Add some variation to the values based on the category
df.loc[df['Category'] == 'A', 'Value'] += 1
df.loc[df['Category'] == 'B', 'Value'] -= 0.5
df.loc[df['Category'] == 'C', 'Value'] += 0.2

# Visualize the distribution of the numerical variable across different categories
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title("Distribution of Numerical Variable by Category")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
heatmap.

```python
# Generate a synthetic dataset with correlated features
np.random.seed(0)
n_samples = 100
n_features = 5

# Create a correlation matrix
corr_matrix = np.array([
    [1, 0.8, 0.2, 0.4, 0.1],
    [0.8, 1, 0.7, 0.3, 0.2],
    [0.2, 0.7, 1, 0.9, 0.5],
    [0.4, 0.3, 0.9, 1, 0.8],
    [0.1, 0.2, 0.5, 0.8, 1]
])

# Generate the dataset using the correlation matrix
data = np.random.multivariate_normal(mean=np.zeros(n_features), cov=corr_matrix, size=n_samples)
df = pd.DataFrame(data, columns=[f'Feature {i+1}' for i in range(n_features)])

# Calculate the correlation matrix of the dataset
corr_matrix_df = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix of the Dataset")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```

 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate x-values from 0 to 10
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)  # Compute the sine of each x-value

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

# Show the plot
output_notebook()  # Use this if you're in a Jupyter Notebook
show(p)
```

 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns.

```python
import pandas as pd
from bokeh.models import ColumnDataSource

# Set random seed for reproducibility
np.random.seed(42)

n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values between 0 and 100
y = np.random.rand(n) * 100  # Random y values between 0 and 100
sizes = np.random.randint(10, 100, size=n)  # Random sizes between 10 and 100
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a new plot
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

p.legend.title = "Colors"
p.legend.location = "top_left"

show(p)
```
